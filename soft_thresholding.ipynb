```python
import numpy as np
import matplotlib.pyplot as plt

def soft_thresholding(x, lambda_val):
    """Soft thresholding operator for Lasso regression"""
    return np.where(np.abs(x) <= lambda_val,
                   0,
                   np.where(x > lambda_val,
                           x - lambda_val,
                           x + lambda_val))

# 
lambda_val = 1.0
x = np.linspace(-3, 3, 1000)

# Soft thresholding
y_soft = soft_thresholding(x, lambda_val)

#  (y = x)
y_original = x

# 
plt.figure(figsize=(10, 6))
plt.plot(x, y_soft, 'r-', linewidth=2.5, label='Soft Thresholding')
plt.plot(x, y_original, 'b--', linewidth=1.5, label='Original')

# 
plt.axvline(x=lambda_val, color='gray', linestyle=':', alpha=0.7)
plt.axvline(x=-lambda_val, color='gray', linestyle=':', alpha=0.7)
plt.text(lambda_val + 0.1, -0.5, r'$\lambda$', fontsize=14)
plt.text(-lambda_val - 0.25, -0.5, r'$-\lambda$', fontsize=14)

# 0
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# 
plt.grid(True, alpha=0.3)
plt.title('SoftThresholding', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend()

plt.tight_layout()
plt.savefig("soft_thresholding.png")
```
