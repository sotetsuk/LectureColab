```python
from numpy.linalg import matrix_rank
from matplotlib import pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
import numpy as np
from sklearn.decomposition import PCA

faces = fetch_olivetti_faces()
X = faces['data']

K = 5  # Number of PCs
mu = np.mean(X, axis=0)
X_centered = X - mu
U, sigma, V_t = np.linalg.svd(X_centered, full_matrices=True)
V_t[:K, :]  # 

for i in range(3):
    plt.figure()
    plt.imshow((V_t[i, :]).reshape(64, 64), cmap='gray')
    plt.show()
```
